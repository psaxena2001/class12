```python
import matplotlib.pyplot as plt  # import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
data = pd.read_csv("Education.csv", usecols=['2017'])
df = pd.DataFrame(data)
minarray=np.array([df['2017']])
# Extract y (frequency) and edges (bins)
y,edges = np.histogram(minarray)
#calculate the midpoint for each bar on the histogram
mid = 0.5*(edges[1:]+ edges[:-1])
df.plot(kind='hist',y='2017',color="g",edgecolor='r',fill=True,hatch='*-')
plt.plot(mid,y,'-^')
plt.title('Girls Education in rural Punjab ')
plt.xlabel('Girls')
plt.show()
```

```python
import matplotlib.pyplot as plt  # creating Histogram with two series of data (year 2010, 2017)
import pandas as pd 
data = pd.read_csv("Education.csv", usecols=['2017','2010'])
df = pd.DataFrame(data)
df.plot(kind='hist',y='2017',title='Year 2017')
df['2010'].plot(kind='hist',title="Year 2010",color='y')
plt.title('Girls education in 2017 and 2010')
plt.xlabel('Girls education in 2010 and 2017')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
